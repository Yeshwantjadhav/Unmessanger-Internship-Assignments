```python
Content=[]

for i in range(1,11):
    url=f"http://quotes.toscrape.com/page/{i}/"
    response=requests.get(url)
    response=response.content
    soup=BeautifulSoup(response,'html.parser')
    li=soup.find('div')
    contents=li.find_all('div',class_='quote')
    for content in contents:
        quotes=content.find('span').text
        author=content.find('small').text
        Content.append([quotes,author])
    
    
```

```python
Content
```

```python
df=pd.DataFrame(Content,columns=['quotes','author'])
```
